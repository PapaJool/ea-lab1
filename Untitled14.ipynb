```python
from collections import defaultdict, deque
import time
import tracemalloc
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import heapq

class Graph:
    def __init__(self, vertices):
        self.graph = defaultdict(list)
        self.vertices = vertices

    def add_edge(self, u, v):
        self.graph[u].append(v)

    def topological_sort_kahn(self):
        in_degree = [0] * self.vertices
        for i in self.graph:
            for j in self.graph[i]:
                in_degree[j] += 1

        queue = []
        for i in range(self.vertices):
            if in_degree[i] == 0:
                queue.append(i)

        cnt = 0

        top_order = []

        while queue:
            u = queue.pop(0)
            top_order.append(u)

            for i in self.graph[u]:
                in_degree[i] -= 1
                if in_degree[i] == 0:
                    queue.append(i)
            cnt += 1

        if cnt != self.vertices:
            print("Граф содержит цикл")
            return []

        return top_order

        from collections import defaultdict, deque

    def topological_sort_kahn_optimized(self):
        in_degree = defaultdict(int)
        for edges in self.graph.values():
            for node in edges:
                in_degree[node] += 1

        zero_in_degree_nodes = deque(node for node in self.graph if in_degree[node] == 0)

        top_order = []
        cnt = 0

        while zero_in_degree_nodes:
            u = zero_in_degree_nodes.popleft()
            top_order.append(u)

            for neighbor in self.graph[u]:
                in_degree[neighbor] -= 1
                if in_degree[neighbor] == 0:
                    zero_in_degree_nodes.append(neighbor)
            cnt += 1

        if cnt != self.vertices:
            print("Граф содержит цикл")
            return []

        return top_order

    @classmethod
    def from_networkx(cls, G):
        num_nodes = len(G.nodes)
        graph = cls(num_nodes)
        for u, v in G.edges:
            graph.add_edge(u, v)
        return graph

def generate_random_dag(num_nodes, num_edges):
    G = nx.gnm_random_graph(num_nodes, num_edges, directed=True)
    while not nx.is_directed_acyclic_graph(G):
        G = nx.gnm_random_graph(num_nodes, num_edges, directed=True)
    return G

def measure_time_and_memory(num_edges, method):
    num_vertices = num_edges + 1  # Число вершин = число рёбер + 1
    graph = Graph.from_networkx(generate_random_dag(num_vertices, num_edges))  # Граф без циклов, каждая вершина соединена с предыдущей

    tracemalloc.start()

    start_time = time.time()
    if method == 'original':
        top_order = graph.topological_sort_kahn()
    elif method == 'optimized':
        top_order = graph.topological_sort_kahn_optimized()
    else:
        raise ValueError("Недопустимый метод")
    end_time = time.time()
    execution_time = end_time - start_time

    peak_memory = tracemalloc.get_traced_memory()[1] / (1024 * 1024)  # Пиковое потребление памяти в МБ
    current_memory = tracemalloc.get_traced_memory()[0] / (1024 * 1024)  # Текущее потребление памяти в МБ

    tracemalloc.stop()
    tracemalloc.clear_traces()

    return execution_time, top_order, current_memory, peak_memory

methods = ['original', 'optimized']
num_edges_list = [10000, 20000, 30000, 40000, 50000, 60000]

# Списки для хранения результатов каждого метода
results_original = {'Number of edges': [], 'Execution time (s)': [], 'Current memory (MB)': [], 'Peak memory (MB)': []}
results_optimized = {'Number of edges': [], 'Execution time (s)': [], 'Current memory (MB)': [], 'Peak memory (MB)': []}

# Заполнение результатов тестов
for num_edges in num_edges_list:
    for method in methods:
        execution_time, _, current_memory, peak_memory = measure_time_and_memory(num_edges, method)
        if method == 'original':
            results_original['Number of edges'].append(num_edges)
            results_original['Execution time (s)'].append(execution_time)
            results_original['Current memory (MB)'].append(current_memory)
            results_original['Peak memory (MB)'].append(peak_memory)
        elif method == 'optimized':
            results_optimized['Number of edges'].append(num_edges)
            results_optimized['Execution time (s)'].append(execution_time)
            results_optimized['Current memory (MB)'].append(current_memory)
            results_optimized['Peak memory (MB)'].append(peak_memory)

# Создание DataFrame для каждого метода
df_original = pd.DataFrame(results_original)
df_optimized = pd.DataFrame(results_optimized)
```
